```python
def plot_P(func, P):
    fig, ax = plt.subplots(1, 2, figsize = (17, 5))
    fig.suptitle(f"{func}. Best score: {np.max(P[1])}")
    T = len(P[1])
    ax[0].plot(np.arange(T), P[1], label = 'Best')
    ax[0].plot(np.arange(T), P[2], label = 'Worst')
    ax[0].plot(np.arange(T), P[3], label = 'Mean')
    ax[0].legend(loc = 'lower right')
    ax[0].set_title('Objective function value')
    
    ax[1].plot(np.arange(T), np.array(P[4]))
    ax[1].set_title('Probabilities')
    plt.show()
    
#     plt.title(f"Population data for {func}")
#     plt.plot(P[1], label = "Best")
#     plt.plot(P[2], label = "Worst")
#     plt.plot(P[3], label = "Mean")
#     plt.xlabel("Iteration")
#     plt.ylabel("Value")
#     plt.legend()
#     plt.show()
```

```python
def  initial_probability_vector():
    p = np.array([0.5 for _ in np.arange(0, d)])
    return p
```

```python
def binary_random(p):
    return int(np.random.rand() < p)
```

```python
def random_individual(p):
    x = np.array([binary_random(p[i]) for i in np.arange(d)])
    return x
```

```python
def random_population(p, N):
    P = np.array([random_individual(p) for _ in np.arange(N)])
    return P
```

```python
def one_max(P):
    return np.sum(P, 1)
```

```python
def dec_one_max(P):
    pop_size = len(P)
    res = np.zeros(pop_size)
    for i in range(pop_size):
        if np.count_nonzero(P[i]) > 0:
            res[i] = np.sum(P[i])
        else:
            res[i] = d + 1
    return res
```

```python
def five_dec_one_max(P):
    pop_size = len(P)
    res = np.zeros(pop_size)
    for i in range(pop_size):
        for segment in range(1, int(d/5) + 1):
            if np.count_nonzero(P[i][segment*5 - 5:segment*5]) > 0:
                res[i] += np.sum(P[i][segment*5 - 5:segment*5])
            else:
                res[i] += 6
    return res
```

```python
def describe(P, fun):
    '''
    Returns:
    (indice of max element in fun(P), mean of fun(P), min of fun(P))
    '''
    evaluated = fun(P)
    return np.argmax(evaluated), np.mean(evaluated), np.min(evaluated)
```

```python
def PBIL(T, N, q1, q2, q3, fun, _d):
    global d
    d = _d
    p = initial_probability_vector()
    P = random_population(p, N)
    solutions = [[] for _ in range(T)]
    best = np.zeros(T)
    worst = np.zeros(T)
    mean = np.zeros(T)
    probs = []
    best_solution = []
    for i in range(T):
        desc = describe(P, fun)
        best_guy = P[desc[0]]
        solutions[i] = best_guy
        best[i] = fun([best_guy])
        mean[i] = desc[1]
        worst[i] = desc[2]
        probs.append(p)
        if best[i] == np.max(best):
            best_solution = best_guy
        p = p * (1-q1) + q1 * best_guy
        p = np.array([(p[k] * (1 - q3) + binary_random(0.5) * q3) if np.random.rand() < q2 else p[k] for k in range(d)])
#         for k in range(d):
#             if np.random.rand() < q2:
#                 p[k] = p[k] * (1 - q3) + binary_random(0.5) * q3
        P = random_population(p, N)
    return best_solution, best, worst, mean, probs
```

```python
def UMDA(T, N, S, fun, _d):
    '''
    Args:
    T: number of iterations
    N: size of population
    S: number of best individuals selected from the population at each iteration
    fun: function we want to maximize
    _d: size of an individual
    '''
    global d
    d = _d
    p = initial_probability_vector()
    P = random_population(p, N)
    best = np.zeros(T)
    worst = np.zeros(T)
    mean = np.zeros(T)
    probs = []
    best_guys = [[] for i in range(T)]
    for i in range(T):
        probs.append(p)
        eval_pop = fun(P)
        eval_pop_argsort = eval_pop.argsort()[::-1]
        indices_of_S_best = eval_pop_argsort[:S]
        p = np.array([np.sum(P[indices_of_S_best][:, i])/S for i in range(d)])
        best[i] = eval_pop[eval_pop_argsort[0]]
        worst[i] = eval_pop[eval_pop_argsort[-1]]
        mean[i] = np.mean(eval_pop)
        best_guys[i] = P[eval_pop_argsort[0]]
        P = random_population(p, N)
        
    return best_guys, best, worst, mean, probs
```

```python
def CGA(T, sigma, fun, _d, N = 2):
    '''
    Args:
    T: number of iterations
    sigma: learning parameter
    fun: function we want to maximize
    _d: size of an individual
    N: size of population
    '''
    global d
    d = _d
    p = initial_probability_vector()
    P = random_population(p, N)
    best = np.zeros(T)
    worst = np.zeros(T)
    mean = np.zeros(T)
    probs = []
    best_guys = [[] for i in range(T)]
    for i in range(T):
        probs.append(p)
        evaluated = fun(P)
        eval_fst_indice, eval_snd_indice = np.argmax(evaluated), np.argmin(evaluated)
        x1, x2 = P[eval_fst_indice], P[eval_snd_indice]
        p = np.array([(p[k] + sigma) if (x1[k] == 1 and x2[k] == 0) 
                      else (p[k] - sigma) if (x1[k] == 0 and x2[k] == 1)
                      else p[k] for k in range(d)])
        best[i], worst[i], mean[i] = np.max(evaluated), np.min(evaluated), np.mean(evaluated)
        best_guys[i] = x1
        P = random_population(p, N)
    return best_guys, best, worst, mean, probs
```

```python
def test_pbil(_d, N, T, theta1, theta2, theta3):
    P = PBIL(T, N, theta1, theta2, theta3, one_max, _d)
    plot_P("one_max", P)

    P = PBIL(T, N, theta1, theta2, theta3, dec_one_max, _d)
    plot_P("dec_one_max", P)

    P = PBIL(T, N, theta1, theta2, theta3, five_dec_one_max, _d)
    plot_P("five_dec_one_max", P)
    
```

```python
def test_umda(_d, N, T, S):
    U = UMDA(T, N, S, one_max, _d)
    plot_P("one_max", U)

    U = UMDA(T, N, S, dec_one_max, _d)
    plot_P("dec_one_max", U)

    U = UMDA(T, N, S, five_dec_one_max, _d)
    plot_P("five_dec_one_max", U)
```

```python
def test_cga(_d, N, T, sigma):
    U = CGA(T, sigma, one_max, _d, N)
    plot_P("one_max", U)

    U = CGA(T, sigma, dec_one_max, _d, N)
    plot_P("dec_one_max", U)

    U = CGA(T, sigma, five_dec_one_max, _d, N)
    plot_P("five_dec_one_max", U)
```
