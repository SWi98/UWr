```python
import numpy as np
import matplotlib.pyplot as plt
import time
import urllib.request
from pymoo.algorithms.nsga2 import NSGA2
from tqdm.auto import tqdm
from pymoo.algorithms.nsga2 import NSGA2
from pymoo.factory import get_problem, get_sampling, get_crossover, get_mutation
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter
from pymoo.model.problem import Problem
import autograd.numpy as anp
%matplotlib inline
```

```python
def run(problem, 
        n_generations, 
        n_iterations, 
        selection = 'tournament', 
        crossover = 'real_ux', 
        mutation = 'bin_bitflip'):
    algorithm = NSGA2(pop_size = n_generations)
    algorithm.setup(problem, termination=('n_gen', n_iterations), seed=1)
    i, plot_cond = 0, n_generations / 10
    plot = Scatter(figsize = (15, 10))
    while algorithm.has_next():
        i += 1
        algorithm.next()
        if i % plot_cond== 0:
            plot.add(algorithm.opt.get('F'), color="blue", alpha=0.7)
    res = algorithm.result()
    plot.add(problem.pareto_front(), plot_type="line", color="black", alpha=0.7)
    plot.add(res.F, color="red")
    plot.show()    
    plt.show()
    return res
```
